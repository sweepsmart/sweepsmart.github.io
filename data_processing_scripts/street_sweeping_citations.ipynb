```python
# Unify the column names for all dataframe based on jan column names
col_uni = ['Tick Badge #', 'Ticket Number', 'Ticket Issue Date',
       'Ticket Issue Time', 'Fine Amount', 'Violation Internal Code',
       'Violation External Code', 'Violation Type', 'Route', 'Location']

uni_months = []
n = 2
for month in all_month:
    print('Month', n)
    uni_mon = month[col_uni]
    uni_months.append(uni_mon)
    display(uni_months[n-2].shape)
    n += 1
```

```python
joined = pd.DataFrame(columns = col_uni)

for month in uni_months:
    joined = joined.append(month, ignore_index = True)
    
display(joined.shape)
joined
```

```python
cleaned_joined = joined.dropna()[col_uni]
cleaned_joined.shape
```

```python
violation_code_mapping = cleaned_joined.groupby(['Violation Internal Code', 'Violation External Code']).agg({'Ticket Number': 'count'})
violation_code_mapping.sort_values(['Ticket Number'], ascending=False)
```
